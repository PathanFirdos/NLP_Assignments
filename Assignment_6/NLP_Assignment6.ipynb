```python
from transformers import MarianMTModel, MarianTokenizer
```

# Load Pretrained Translation Model

```python
model_name = "Helsinki-NLP/opus-mt-en-hi"

tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

# Translation Function

```python
def translate_en_to_hi(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**inputs)
    return tokenizer.decode(translated[0], skip_special_tokens=True)
```

# Translate Public Information Content

```python
text = "The government has announced a new healthcare policy for senior citizens."

translation = translate_en_to_hi(text)
print("English :", text)
print("Hindi   :", translation)
```

# Multiple Sentences Translation

```python
texts = [
    "India launched a new digital education program.",
    "Citizens are advised to follow safety guidelines.",
    "Weather alerts have been issued for coastal areas."
]

for t in texts:
    print("EN:", t)
    print("HI:", translate_en_to_hi(t))
    print()
```
