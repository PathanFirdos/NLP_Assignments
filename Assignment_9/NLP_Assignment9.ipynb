```python
from transformers import pipeline
```

```python
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-xlm-roberta-base-sentiment"
)
```

```python
feedback_data = [
    "सरकार की यह योजना बहुत अच्छी है और लोगों को फायदा हो रहा है",
    "बस सेवा बहुत खराब है, समय पर नहीं आती",
    "अस्पताल की सुविधा ठीक थी लेकिन डॉक्टर देर से आए",
    "ऑनलाइन आवेदन प्रक्रिया बहुत आसान है",
    "ग्राहक सेवा से मैं बिल्कुल संतुष्ट नहीं हूँ"
]
```

```python
results = []

for text in feedback_data:
    output = sentiment_analyzer(text)[0]
    results.append((text, output["label"], output["score"]))

results
```

```python
import pandas as pd

df = pd.DataFrame(results, columns=["Feedback", "Sentiment", "Confidence"])
df
```
